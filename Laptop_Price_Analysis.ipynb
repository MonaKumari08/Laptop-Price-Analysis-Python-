**Extracting the X Resolution and the Y Resolution**

```python
df_split = df['ScreenResolution'].str.split('x',n = 1,expand=True)

df['X_res'] = df_split[0]
df['Y_res'] = df_split[1]
df.sample(5)
```

```python
# As we see X_res consists of whole textand we need to extract the digits from it.
# Firstly replace all the "," with "" and then find all numbers from that string as
# "\d+.?\d+",\d means that integer number and
#.? all the numbers which come after an number and
# \d+ the string must end with number


df['X_res'] = df['X_res'].str.replace(',','').str.findall(r'(\d+.?\d+)').apply(lambda x:x[0])
df.head()
```

```python
df.info()
```

```python
df['X_res'] = df['X_res'].astype('int')
df['Y_res'] = df['Y_res'].astype('int')
df.info()
```

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10,5))
sns.heatmap(numeric_df.corr(), annot=True,cmap='plasma')
```

```python

print(numeric_df.corr()['Price'])

# As the X_res and Y_res is increasing,the price of the laptop is also increasing,
# so X_res and Y_res are positively correlated.
```

**Creating PPI column in the dataset**

```python
# We can create a new column named PPI{pixels per inch},now as we saw from the
# correlation plot that the X_res and Y_res are having much collinearity,
# so we will combine them with Inches which is having less collinearity,


df['PPI'] = (((df['X_res']**2) + (df['Y_res']**2))**0.5/df['Inches']).astype('float')
df.head()
```

```python
numeric_df.corr()['Price']
```

```python
# PPI is having good correlation,so we will be using that,
# as that is a combination of 3 features and that gives collective results of 3 columns,
# so we will drop Inches,X_res,Y_res as well

df.drop(columns=['ScreenResolution','Inches','X_res','Y_res'],inplace=True)
df.head()
```

### **Analysis on the memory column**

```python
df['Memory'].value_counts()
```

```python
# 4 most common variants observed : HDD, SDD, Flash Storage, Hybrid
# We will remove decimal space for example 1.0 TB will become 1TB
# We will replace the word GB with ""
# We will replace the word TB with "000"
# We will split the word across the "+" character
```

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB','')
df['Memory'] = df['Memory'].str.replace('TB','000')
newdf = df['Memory'].str.split('+', n=1, expand=True)
newdf
```

```python
# Eliminating white spaces

df['First'] = newdf[0].str.strip()
df.sample(10)
```

```python
#Creating Layer 1 column for each memory type

def applychanges(value):

    df['Layer1'+value] = df['First'].apply(lambda x:1 if value in x else 0)


listtoapply = ['HDD','SSD','Hybrid','FlashStorage']
for value in listtoapply:
    applychanges(value)


df.head()
```

```python
# Remove all the character and just keeping the numbers

df['First'] = df['First'].str.replace(r'\D','', regex = True)
df['First'].value_counts()
```

```python
df['Second'] = newdf[1]
df.head()
```

```python
# Creating Layer 2 column for each memory type

def applychanges1(value):
    df['Layer2'+value] = df['Second'].apply(lambda x:1 if value in x else 0)

list_to_apply1 = ['HDD', 'SSD', 'Hybrid', 'FlashStorage']
df['Second'] = df['Second'].fillna("0")
for value in list_to_apply1:
    applychanges1(value)

df.head()
```

```python
# Remove all the characters and just keeping the numbers again

df['Second'] = df['Second'].str.replace(r'\D','',regex = True)
df['Second'].value_counts()
```

```python
df.info()
```

```python
df['First'] = df['First'].astype('int')
df['Second'] = df['Second'].astype('int')
df.info()
```

```python
# multiplying the elements and storing the result in subsequent columns

df["HDD"]=(df["First"]*df["Layer1HDD"]+df["Second"]*df["Layer2HDD"])
df["SSD"]=(df["First"]*df["Layer1SSD"]+df["Second"]*df["Layer2SSD"])
df["Hybrid"]=(df["First"]*df["Layer1Hybrid"]+df["Second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["First"]*df["Layer1FlashStorage"]+df["Second"]*df["Layer2FlashStorage"])
```

```python
# dropping unnecessary columns

df.drop(columns=['First', 'Second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                'Layer1FlashStorage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                'Layer2FlashStorage'], inplace=True)
```

```python
df.sample(5)
```

```python
df.drop(columns=['Memory'],inplace=True)
df.sample(5)
```

```python
numeric_df = df.select_dtypes(include='number')
numeric_df.corr()['Price']
```

```python
# Based on the correlation we observe that Hybrid and Flash Storage are almost negligible,so we can simply drop them off,
# where as HDD and SDD are having good correlation,we find that HDD has -ve relation with Price,and
# if the price of laptop is increasing there is more probability that the laptop is going to use SDD instead of HDD and vice versa as well.

df.drop(columns = ['Hybrid','Flash_Storage'],inplace=True)
df.head()
```

## **Operating System analysis**

```python
df['OpSys'].value_counts()
```

**Price variation on OS in laptops**

```python
sns.barplot(x = df['OpSys'], y = df['Price'], palette= 'winter')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# There are many versions of one type of OS, espicially for Windows and macOS. So we can club them together like,

# club {Windows 10,Windows 7,Windows 7 S}-->Windows
# club {macOS,mac OS X}--> mac
# others --> others

def OScategory(text):

    if text == 'Windows 10' or text == 'Windows 7' or text == 'Windows 10 S':
        return 'Windows'
    elif text == 'Mac OS X' or text == 'macOS':
        return 'Mac'
    else:
        return 'Other'

df['OpSys'] = df['OpSys'].apply(lambda x : OScategory(x))
df.head(1000)
```
